## K-means clustering

```python
# import thư viện
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
```

## Created a Dataset

```python
centers = np.array([
    [-5, -5],
    [-5, 5],
    [5, -5],
    [5, 5],
])

blobs, groups = make_blobs(n_samples=200, n_features=2, centers = centers, cluster_std = 1.5)
```

```python
# Draw the blobs
plt.scatter(blobs[:, 0], blobs[:, 1], c='gray')
plt.show()
```

## K-Means Clustering Algorithm

```python
# Create random centroids
# Centroids là các điểm trung tâm
colors = ['r', 'y', 'g', 'b']
centroids = np.array([
    [-7, -2],
    [10, 10],
    [-3, 12],
    [0, 0],
])

# plot the centroids
for i in range(4):
    plt.scatter(centroids[i:i+1, 0], centroids[i: i + 1, 1], c = colors[i], s = 100)
plt.scatter(blobs[:, 0], blobs[:, 1], c='gray')
plt.show()
```

```python
for _ in range(200):
    # Phân bổ các điểm dựa trên trọng tâm
    allocations = []

    for j in range(blobs.shape[0]):
        distances = np.sum((blobs[j] - centroids) ** 2, axis = 1)
        argmin = np.argmin(distances)
        allocations.append(argmin)

    # Chuyển sang numpy
    allocations = np.array(allocations)
    
    # Tính toán các điểm trọng tâm 
    for j in range(centroids.shape[0]):
        centroids[j] = blobs[allocations == j, :].mean(axis=0)
```

```python
# Plot the centroids
for i in range(4):
    plt.scatter(centroids[i:i+1, 0], centroids[i: i + 1, 1], c = colors[i], s = 300)
plt.scatter(blobs[:, 0], blobs[:, 1], c='gray')
plt.show()
```
